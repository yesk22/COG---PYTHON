```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# PARAMETROS DE ENTRADA :
# Colocar coordenadas
cx = np.array([[1,2],[5,3],[5,1],[9,4]])
# Colocar tasa de transporte
t = np.array([900,150,625,100])
# discretización de malla
n = 0.01
```

```python
## DEFINIENDO MAPA
mx = max(cx[:,0])
my = max(cx[:,1])
lx = round(mx+1)            # maximo valor de la coordenada en y
ly = round(my+1)            # maximo valor de la coordenada en x
vx = np.arange(0,lx,n)      # vector con coordenadas en x
vy = np.arange(0,ly,n)      # vector con coordenadas en y
X, Y = np.meshgrid(vx, vy)  # Coordenadas de mapa (x,y)
```

```python
## CALCULANDO COSTOS PARA CADA REGIÓN
Z1 = t[0]*np.sqrt(np.power(X-cx[0,0],2) + np.power(Y-cx[0,1],2))*100;
Z2 = t[1]*np.sqrt(np.power(X-cx[1,0],2) + np.power(Y-cx[1,1],2))*100;
Z3 = t[2]*np.sqrt(np.power(X-cx[2,0],2) + np.power(Y-cx[2,1],2))*100;
Z4 = t[3]*np.sqrt(np.power(X-cx[3,0],2) + np.power(Y-cx[3,1],2))*100;
Z = Z1 + Z2 + Z3 + Z4;
ind = np.unravel_index(np.argmin(Z, axis=None), Z.shape)              # costo minimo de transporte
print(ind)
```

```python
## PLOT DE MAPA CON REGIONES
plt.show()
fig1, ax2 = plt.subplots(layout='constrained')
plt.grid()
CS = ax2.contourf(X, Y, Z, 20, cmap=plt.cm.YlOrBr)
ax2.set_title('Region map with locations')
ax2.set_xlabel('x coordinates')
ax2.set_ylabel('y coordinates')

plt.plot(cx[0,0],cx[0,1], marker="o", color="black")
plt.plot(cx[1,0],cx[1,1], marker="o", color="black")
plt.plot(cx[2,0],cx[2,1], marker="o", color="black")
plt.plot(cx[3,0],cx[3,1], marker="o", color="black")
plt.plot(ind[0]/100,ind[1]/100, marker="o", color="green")
plt.text(ind[0]/100,ind[1]/100, "Valor mínimo",bbox=dict(facecolor='green', alpha=0.5))
```
